```python
import google.generativeai as genai
import PIL.Image
import time
import os

# 1. إعداد المفتاح
MY_API_KEY = "AIzaSyDDfhViSnKSMAl5VSJVg7kCNqWSEIAKfRQ"
genai.configure(api_key=MY_API_KEY)

def get_best_model():
    """
    دالة ذكية تبحث عن الموديل المتاح في حسابك
    """
    print("🔍 جاري البحث عن الموديلات المتاحة لحسابك...")
    available_models = []
    try:
        for m in genai.list_models():
            if 'generateContent' in m.supported_generation_methods:
                available_models.append(m.name)

        print(f"📋 الموديلات التي وجدتها: {available_models}")

        # ترتيب الأفضلية: نفضل فلاش 1.5، ثم برو 1.5، ثم أي شيء آخر
        priority_list = [
            'models/gemini-1.5-flash',
            'models/gemini-1.5-flash-001',
            'models/gemini-1.5-flash-latest',
            'models/gemini-1.5-pro',
            'models/gemini-1.5-pro-001',
            'models/gemini-pro' # هذا لا يدعم الصوت، لكنه بديل أخير
        ]

        for model_name in priority_list:
            if model_name in available_models:
                # نحذف كلمة models/ من البداية لأن الدالة generate_content لا تحبها أحياناً
                clean_name = model_name.replace("models/", "")
                print(f"✅ تم اختيار الموديل: {clean_name}")
                return genai.GenerativeModel(clean_name)

        # إذا لم نجد أي شيء من القائمة، نأخذ أول واحد متاح وخلاص
        if available_models:
            first_model = available_models[0].replace("models/", "")
            print(f"⚠️ لم نجد الموديلات المفضلة، سنستخدم: {first_model}")
            return genai.GenerativeModel(first_model)

    except Exception as e:
        print(f"❌ خطأ في البحث عن الموديلات: {e}")
        return None

# 2. تهيئة الموديل المختار
model = get_best_model()

def smart_chat(user_text=None, image_path=None, audio_path=None):
    if not model:
        return "❌ فشل تهيئة الموديل. تأكد من المفتاح."

    prompt_parts = []

    if user_text: prompt_parts.append(user_text)

    # معالجة الصور
    if image_path and os.path.exists(image_path):
        prompt_parts.append(PIL.Image.open(image_path))
        print("📷 صورة مرفقة.")

    # معالجة الصوت
    if audio_path and os.path.exists(audio_path):
        print("🎙️ جاري رفع الصوت...")
        myfile = genai.upload_file(audio_path)
        while myfile.state.name == "PROCESSING":
            time.sleep(1)
            myfile = genai.get_file(myfile.name)
        prompt_parts.append(myfile)
        print("✅ الصوت جاهز.")

    try:
        response = model.generate_content(prompt_parts)
        return response.text
    except Exception as e:
        return f"خطأ: {e}"

# --- التشغيل ---
if model:
    print("\n💬 سأجرب سؤالاً نصياً بسيطاً للتأكد:")
    print(smart_chat(user_text="أهلاً، هل أنت جاهز للعمل؟"))
else:
    print("لا يمكن بدء الشات.")
```

```python
import google.generativeai as genai
import PIL.Image
import time
import os

# ==========================================
# 1. إعدادات المشروع
# ==========================================
MY_API_KEY = "AIzaSyDDfhViSnKSMAl5VSJVg7kCNqWSEIAKfRQ" # مفتاحك الذي يعمل
MODEL_NAME = "gemini-2.5-flash"  # اخترنا هذا الموديل من قائمتك لأنه الأسرع والأحدث

genai.configure(api_key=MY_API_KEY)

# ==========================================
# 2. كلاس الشات بوت (The Brain)
# ==========================================
class GeminiChatBot:
    def __init__(self):
        # تهيئة الموديل
        self.model = genai.GenerativeModel(MODEL_NAME)
        # بدء جلسة محادثة فارغة (لحفظ التاريخ)
        self.chat_session = self.model.start_chat(history=[])
        print(f"🤖 تم تشغيل البوت باستخدام موديل: {MODEL_NAME}")

    def upload_audio(self, audio_path):
        """دالة مساعدة لرفع الملفات الصوتية وانتظار معالجتها"""
        if not os.path.exists(audio_path):
            return None

        print(f"🎙️ جاري رفع ومعالجة الملف الصوتي: {audio_path}...")
        audio_file = genai.upload_file(path=audio_path)

        # حلقة انتظار حتى تنتهي جوجل من معالجة الملف
        while audio_file.state.name == "PROCESSING":
            time.sleep(1)
            audio_file = genai.get_file(audio_file.name)

        if audio_file.state.name == "FAILED":
            print("❌ فشل في معالجة الملف الصوتي")
            return None

        print("✅ الملف الصوتي جاهز.")
        return audio_file

    def send_message(self, text=None, image_path=None, audio_path=None):
        """
        الدالة الرئيسية التي تستقبل أي نوع من المدخلات وترد عليها
        """
        input_parts = []

        # 1. تجهيز النص
        if text:
            input_parts.append(text)

        # 2. تجهيز الصورة (إن وجدت)
        if image_path:
            try:
                img = PIL.Image.open(image_path)
                input_parts.append(img)
                print("📷 تم إرفاق صورة.")
            except Exception as e:
                return f"❌ خطأ في الصورة: {e}"

        # 3. تجهيز الصوت (إن وجد)
        if audio_path:
            audio_file = self.upload_audio(audio_path)
            if audio_file:
                input_parts.append(audio_file)
            else:
                return "❌ تعذر رفع الملف الصوتي."

        # إذا لم يكن هناك مدخلات
        if not input_parts:
            return "⚠️ لا يوجد مدخلات للإرسال!"

        # 4. الإرسال للموديل والحصول على الرد
        try:
            # نستخدم chat_session لإرسال الرسالة وحفظ السياق
            response = self.chat_session.send_message(input_parts)
            return response.text
        except Exception as e:
            return f"❌ خطأ في النظام: {e}"

    def get_history(self):
        """لعرض تاريخ المحادثة بالكامل"""
        return self.chat_session.history

# ==========================================
# 3. منطقة التجربة (Run & Test)
# ==========================================

# إنشاء نسخة من البوت
my_bot = GeminiChatBot()

print("\n--- 💬 بدء المحادثة ---")

# تجربة 1: نص عادي (سيتذكر البوت اسمي لاحقاً)
response1 = my_bot.send_message(text="أهلاً، أنا اسمي محمد وأدرس علوم حاسب.")
print(f"Gemini: {response1}")

# تجربة 2: سؤال يعتمد على الذاكرة
response2 = my_bot.send_message(text="ما هو تخصصي الذي ذكرته لك للتو؟")
print(f"Gemini: {response2}")

# تجربة 3: (اختياري) صورة + نص
# response3 = my_bot.send_message(text="اشرح لي الكود في الصورة", image_path="code.png")
# print(f"Gemini: {response3}")

# تجربة 4: (اختياري) صوت
# response4 = my_bot.send_message(text="لخص هذا الاجتماع", audio_path="meeting.mp3")
# print(f"Gemini: {response4}")
```
